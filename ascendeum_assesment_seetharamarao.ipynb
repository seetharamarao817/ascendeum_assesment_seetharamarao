```python
def matrix_manipulation(n):
    # Generate Numbers 
    total_values = (n*n)+1
    numbers_list = [i for i in range(1,total_values)]
     
    print(f"Generated Numbers: {numbers_list}")
    # empty N x N matrix creation 
    
    empty_m = [[0 for _ in range(n)] for _ in range(n)]
    
    print(f"Created empty NxN matrix")
    
    for i,j in enumerate(empty_m):
        print(j,end = ' ')
        print()
    
    top, bottom, left, right = 0, n-1, 0 , n-1 
    counter = 0 
    while top <= bottom and left <= right:
        
        # First we will start by filling top left to right 
        
        for i in range(left,right+1):
            empty_m[top][i] = numbers_list[counter]
            counter+= 1
        top += 1
            
        
        # Next we will start by filling from right top to bottom 
        for j in range(top,bottom+1):
            empty_m[j][right] = numbers_list[counter]
            counter +=1
        right -=1 
        
        # Next we will fill the values from bottom right to left 
        if top <= bottom:
            for k in range(right,left-1,-1):
                empty_m[bottom][k] = numbers_list[counter]
                counter +=1
            bottom -=1 
            
        if left<=right:
            for l in range(bottom,top-1,-1):
                empty_m[l][left] = numbers_list[counter]
                counter +=1
            left +=1
    print(f"Final Filled NxN matrix")

    for i,j in enumerate(empty_m):
        print(j,end = ' ')
        print()
    return empty_m
            
        
            
    
    
```
